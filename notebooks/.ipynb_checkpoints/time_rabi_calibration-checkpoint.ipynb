```python
trabi.run_and_report(jm, 1000, qchip, fit_type='period', period=)
```

```python
shots = trabi.shots
avg_response = np.average(shots['Q2'], axis=1)
```

```python
plt.plot(pulse_width_interval, avg_response)
```

```python
freq_ind_max = np.argmax(np.abs(np.fft.rfft(avg_response)[1:])) + 1
print(np.fft.rfft(avg_response))
```

```python
freq_max = np.fft.rfftfreq(len(avg_response), np.diff(pulse_width_interval)[0])[freq_ind_max]
```

```python
plt.plot(pulse_width_interval, avg_response)
plt.plot(pulse_width_interval, trabi._cos(pulse_width_interval, -0.5, 0.5, 1/freq_max, 0))
```
